# $zip()$
Die $zip()$-Funktion verbindet mehrere iterierbare Objekte und erzeugt ein Tuple mit den Inhalten aus den Objekten.

```python
help(zip)
```

```python
a = range(8, -1, -1)
b = range(20,26)
c = 'Hallo'

for i in zip(a,b, c):
    print(i)
```

```python
a = range(0,5)
b = range(20,25)
n,m,g = [4,6,8]
print(n,m,g)
for i,j in zip(a,b):
    print('i={} & j={}'.format(i,j))
```

# $enumerate()$
Die $enumerate()$-Funktion zählt die Iterationsschritte.

```python
x = ['Steffen','Arthur','Marco','Sonja','Simon','Karin','Till']
x
```

```python
for i,name in enumerate(x):
    print(i,':'*i + name)
```

Hinweise für die Verwendung von Funktionen und Methoden aus importierten Bibliotheken

```python
def bp_some_func(x, y, z=3.14, **kwargs):
    r"""Some function does some stuff. Parameters:
    ----------
    x : int
        Description of x
    y : str
        Description of y
    z : float, optional
        Description of z.  Defaults to 3.14
    **kwargs
        Arbitrary optional keyword arguments.
        w : float
           Defaults to 6.28    
    """
    w = kwargs.pop("w", 6.28)
    if kwargs:
        print("Got {0} unused kwargs".format(len(kwargs)))
    return (x**2 + len(y)) * (w + z)
```

### 3. $if$, $elif$ und $else$ Anweisungen
$if$ und $elif$ werden mit Bedingungen versehen, $else$ macht den Rest.

```python
name = input('Zahl bitte! >>>  ')
zahl = float(name)

if type(zahl) != float:
    print('Error')
else:
    zahl = zahl**4
    print('Zahl: {}'.format(zahl))
```

```python
a = np.random.randint(0,100,1)

if a < 33:
    print('Erstes Drittel')
elif a < 66:
    print('Zweites Drittel')
else: 
    print('Drittes Drittel')
print(a)
```

# Übungsaufgaben

```python
# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares and green triangles
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```
