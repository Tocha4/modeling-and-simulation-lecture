```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def velocity(K,nu,dp,rho,dx):
    
    u = -K/nu*dp/(np.sum(rho)*dx)
    return u
```

```python
K = 1e-11 # m²
nu = 0.1 # mPas
dp = 800 # Pa

length = 0.2 # m
time = 120 # sec

dx = 0.1 # m
dt = 0.1 # sec

rho = np.zeros((int(time/dt), int(length/dx))) # Normierte Füllmenge 
rho[:,0] = 1 # Randbedingung
```

```python
for n in range(len(rho[:,0])):
    u = velocity(K,nu,dp,rho,dx)
    stability = u*dt/dx
    print(stability)
    for i in range(1,len(rho[0,:])-1):
        
        rho[n+1,i] = rho[n,i] + stability*(rho[n,i]-rho(n,i-1))
```
