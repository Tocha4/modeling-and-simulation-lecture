# Simulation der Aushärtung von duromeren Systemen

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<center><h5> Modell nach Kamal und Sourour

\begin{equation*}
\dfrac{\partial\alpha}{\partial t} = \left( k_{1} + k_{2}\cdot\alpha^{m}\right)\left(1-\alpha \right)^n 
\end{equation*}

```python
def model(a,t,k,k2,n,m):
    
    dadt = (k+k2*a**m)*(1-a)**n
    
    return dadt
```

##### Reaktionskonstanten nach Arrhenius

```python
def k1_f(T):
    a = 4.33
    E = 24.1
    R = 8.31448/1000
    k = a*np.exp(-E/(R*T))
    return k

def k2_f(T):
    a = 1.41e+10
    E = 64.3
    R = 8.31448/1000
    k = a*np.exp(-E/(R*T))
    return k

n = 0.99
m = 3.56
```

##### Lösung des Integrals 

\begin{equation*}
\alpha = \int_{t_0}^{t} \dfrac{\dot{q}}{q_T} dt
\end{equation*}

```python
from scipy.integrate import odeint

T = 290 # Temperatur in Kelvin
a0 = 0 # Lösung bei a(t0) --> Anfangsbedingung
t = np.linspace(0,1000,1000) # Zeitraum in sec
a = odeint(model, a0, t, args=(k1_f(T), k2_f(T), 3.56, 0.99)) # Integration des Modells
dadt = np.gradient(a[:,0], t) # Reaktionsgeschwindigkeit [1/sec]
```

```python
def plot_a_dadt(t,a,dadt, label=None):
    plt.subplot(2,1,1)
    plt.plot(t,a, label='{} K'.format(label))
    plt.ylabel(r'Reaktionsumsatz $\alpha$', fontsize=12)
    plt.xlabel(r'Time $[sec]$', fontsize=12)
    if label != None:
        plt.legend()

    plt.subplot(2,1,2)
    plt.plot(t,dadt, label='{} K'.format(label))
    plt.ylabel(r'Reaktionsgeschwindigkeit $\dfrac{\partial\alpha}{\partial t}$ $[sec^{-1}]$', fontsize=12)
    plt.xlabel(r'Time $[sec]$', fontsize=12)
    if label != None:
        plt.legend()

plt.figure(figsize=(12,8))
plot_a_dadt(t,a,dadt)
plt.show()
```

##### Vergleich der Aushärtung bei unterschiedlichen konstanten Temperaturen

```python
t = np.linspace(0,600,1000)
plt.figure(figsize=(14,10))

for T in range(300,270,-5): # Iteration durch unterschiedliche Temperaturen
    a = odeint(model, a0, t, args=(k1_f(T), k2_f(T), 3.56, 0.99))
       
    dadt = np.gradient(a[:,0], t)
    
    plot_a_dadt(t,a,dadt,T)

plt.show()
```

# Miteinbeziehung der lokalen Temperatur

\begin{equation*}
\dfrac{\partial T}{\partial t}  =  \dfrac{\dfrac{\partial\alpha}{\partial t} \cdot q_T}{\rho \cdot c_p} 
\end{equation*}

```python
mat1 = 'Duromer'
material = {'Duromer': [0.2, 1.2, 1.5]} # Wärmeleitfähigkeit, Dichte, spez. Wärmekapazität
mat = material[mat1]
time = 300 # Zeit [sec]
dt = 0.25 # Zeitschritt [sec]
Hr = 2 # Reaktionsenthalpie [J/mol]
n = 1 # Stoffmenge [mol]
qt = Hr*n # Gesamte freigesetzte Wärme [J]
Cq = 4*dt/(mat[1]*mat[2])
```

```python
#     plt.subplot(2,1,1)
#     plt.plot(t,a, label='{} K'.format(T))
#     plt.ylabel(r'Reaktionsumsatz $\alpha$', fontsize=12)
#     plt.xlabel('Time [sec]', fontsize=12)
#     plt.legend()
    
#     plt.subplot(2,1,2)
#     plt.plot(t,dadt, label='{} K'.format(T))
#     plt.ylabel(r'Reaktionsgeschwindigkeit $\dfrac{\partial\alpha}{\partial t}$ $[sec^{-1}]$', fontsize=12)
#     plt.xlabel('Time [sec]', fontsize=12)
#     plt.legend()plt.figure(figsize=(14,10))
for T_s in range(270,300,5):
    
    T = np.zeros(int(time/dt))
    T[0] = T_s
    a = [0]
    dt_a = np.linspace(0,dt,100)
    
    dadt_points = []
    a_points = []
    for n in range(len(T)-1):
        
        a = odeint(model, a[-1], dt_a, args=(k1_f(T[n]), k2_f(T[n]), 3.56, 0.99))
        dadt = np.gradient(a[:,0], dt_a)
        dadt_points.append(dadt[-1])
        a_points.append(a[-1])
        
        T[n+1] = dadt[-1]*Cq + T[n]
    
    plt.subplot(2,1,1)
    t = np.linspace(0,time,len(T)-1)
    plt.plot(t, a_points, '--', label=T_s)
    plt.ylabel(r'Reaktionsumsatz $\alpha$', fontsize=12)
    plt.xlabel('Time [sec]', fontsize=12)
    plt.legend()
    
    plt.subplot(2,1,2)
    plt.plot(t, dadt_points, '--',label=T_s)
    plt.ylabel(r'Reaktionsgeschwindigkeit $\dfrac{\partial\alpha}{\partial t}$ $[sec^{-1}]$', fontsize=12)
    plt.xlabel('Time [sec]', fontsize=12)
    plt.legend()
plt.show()
```
